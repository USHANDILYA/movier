***Converting Data to efficient format***

*First we will extract genre names from the list of dictionaries in genre column.Initially the list is present as string of list so we convert it into list format using a module "ast".Then we extract the genre names*

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head(1)
```

*Applying same method to keywords*

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head(1)
```

*Now we extract names of top 3 cast members*

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head(1)
```

*Now we extract only those names from crew whose job is director*

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head(1)
```

```python
#converting overview into list
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head(1)
```

*We remove blankspaces to avoid confusion for our model in between names* 

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head(1)
```

*Now we will combine all the list and convert them into string to create tags column*

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head(1)
```

```python
new_df=movies[['movie_id','title','tags']]
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
new_df.head(1)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

We have few similar words which can be a problem in our model.Like 'action' and 'actions' both are not needed.So we apply stemming using nltk.

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y= []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

**Text Vectorisation**

To calculate similarity between 2 movies,we need to calculate the similarity score based on tags column of both movies.For this we use vectorisation.Vectors can be assumed as coordinates in the x-y plane.So for a given movie vector,we calculate the closest vectors/coordinates based on the x-y coordinates.We are using *Bag of words* approach.

```python
from sklearn.feature_extraction .text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()#explicitly converting to numpy array
vectors
```

```python
cv.get_feature_names() #5000 most used words in corpus
```

Now we calculate cosine distance(angle) of each movie with every other movie in vector space.Greater the distance ,lesser the similarity.(Euclidean distance i.e ((x2-x1)^2+(y2-y1)^2)^1/2 fails on larger dataset or high dimension so cosine is used)

```python
from sklearn.metrics.pairwise import cosine_similarity
sim=cosine_similarity(vectors)
```

```python
sim.shape
```

**We create a function to recommend 5 similar movies based on similarity score**

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=sim[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
